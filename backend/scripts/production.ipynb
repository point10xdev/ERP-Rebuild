```python


xls = pd.ExcelFile("PhD Data Departments Wise (Updated).xlsx")
sheet_names = xls.sheet_names
all_sheets = {sheet: xls.parse(sheet) for sheet in sheet_names}

manual_replacements = {
    's.no': 'S.No',
    "student's name": "Student's Name",
    'department': 'Department',
    'department ': 'Department',
    'enrolment no.': 'Enrolment No',
    'enrolment no': 'Enrolment No',
    'registration no': 'Registration No',
    'batch': 'Batch',
    'batch ': 'Batch',
    'gender': 'Gender',
    'gender ': 'Gender',
    'date of joining': 'Date of Joining',
    'admission category': 'Admission Category',
    'full time/part time': 'Full Time/Part Time',
    'full time/part time ': 'Full Time/Part Time',
    'full time/part time.1': 'Full Time/Part Time',
    'full time/part-time': 'Full Time/Part Time',
    'full time/part-time.1': 'Full Time/Part Time',
    'full time ': 'Full Time/Part Time',
    'full time': 'Full Time/Part Time',
    'full time/part-time': 'Full Time/Part Time',
    'full time/part-time.1': 'Full Time/Part Time',
    'remark': 'Remark',
    'remark ': 'Remark',
    'supervisor': 'Supervisor',
    'co-supervisor ': 'Co-supervisor',
    'Co-supervisor': 'Co-supervisor',
    'designation of co-supervisor': 'Designation of Co-Supervisor',
    'designation of co-supervisor ': 'Designation of Co-Supervisor',
    'designation of supervisor .1': 'Designation of Supervisor',
    'designation of supervisor ': 'Designation of Supervisor',
    'Designation of Supervisor': 'Designation of Supervisor',
}
column_names = set(manual_replacements.values())
combined_df = pd.DataFrame(columns=sorted(column_names))
def normalize(text):
    text = unicodedata.normalize("NFKD", text)
    text = text.replace('â€“', '-').replace('â€”', '-')  # normalize en-dash/em-dash to hyphen
    text = re.sub(r'[^\x00-\x7F]+', '', text)        # remove non-ASCII characters
    text = re.sub(r'\s+', ' ', text)                 # collapse all whitespace
    return text.strip().lower()


# Normalize all keys in manual_replacements
manual_replacements_normalized = {
    normalize(k): v for k, v in manual_replacements.items()
}
def normalize_col(col):
    cleaned = normalize(col)
    return manual_replacements_normalized.get(cleaned, None)

for sheet, df in all_sheets.items():
    cleaned_df = pd.DataFrame(columns=combined_df.columns)
    for col in df.columns:
        new_col = normalize_col(col)
        if new_col:
            cleaned_df[new_col] = df[col]
    combined_df = pd.concat([combined_df, cleaned_df], ignore_index=True)
def parse_date(value):
    try:
        # Handle Excel serial numbers (e.g., 42247)
        if isinstance(value, (int, float)):
            return pd.to_datetime(value, unit='D', origin='1899-12-30')
        # Handle string dates (e.g., "8/31/2015")
        elif isinstance(value, str):
            # Remove non-numeric characters (e.g., hidden whitespace)
            cleaned_value = re.sub(r'[^0-9/]', '', value)
            return pd.to_datetime(cleaned_value, format='%m/%d/%Y', errors='coerce')
        else:
            return pd.NaT
    except:
        return pd.NaT
columns_to_drop = [
    'Designation of Supervisor', 
    'Remark',
    'Designation of Co-Supervisor',
    'S.No'
]
combined_df = combined_df.drop(columns=columns_to_drop, errors='ignore')
# Remove extra spaces from all string entries across the DataFrame
combined_df = combined_df.applymap(lambda x: ' '.join(x.split()) if isinstance(x, str) else x)
combined_df = combined_df.replace('Nil', '')
# for col in combined_df.columns:
#     print(f"\nUnique values in '{col}':")
#     print(combined_df[col].unique())

# print(combined_df['Department'].unique())

combined_df.to_excel("PhD_Combined_Cleaned.xlsx", index=False)
```
